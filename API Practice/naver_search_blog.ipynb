```python
### 블로그 검색 연습
# 네이버 웹툰에 대한 블로그 글을 검색하여 기존 네이버 웹툰 데이터베이스에 합친다.
import os
import sys
import urllib.request


client_id = "{MY_CLIENT_ID}"
client_secret = "{MY_SECRET_KEY}"

# 10,20,30대 남성 여성 들이 많이 보는 웹툰 5위 웹툰들 리스트
webtoonList = ['싸움독학','투신전생기','입학용병','약한영웅','나만보여','맘마미안','천하제일인','별이삼샵','판사 이한영','이번 생도 잘 부탁해','곱게키웠더니 짐승','소녀재판','오로지 너를 이기고 싶어','살아남은 로맨스','동생친구','닥터앤 닥터 육아일기','힙한남자','반드시 해피엔딩','아홉수 우리들']
resultList = []
resultDict = {}

for webtoonName in webtoonList:
    # 웹툰 싸움독학 이런식으로 검색
    encText = urllib.parse.quote("웹툰 "+webtoonName)
    url = "https://openapi.naver.com/v1/search/blog?query=" + encText # json 결과
    # url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200): # page ok
        response_body = response.read()
        #print(response_body.decode('utf-8'))
        resultList.append(response_body.decode('utf-8'))
        #검색 결과가 검색날짜, 총 개수, 시작점, 보여주는 개수, 아이템 이렇게 딕셔너리 형태,
        #아이템의 밸류값은 리스트안에 딕셔너리로 블로그 글 제모그 링크, 설명, 블로그 이름 등을 표현
        #사실 아이템 값이 데이터 분석에 중요하므로 이 아이템 값만 빼오는 것이 목적
    else:
        print("Error Code:" + rescode)
```

```python
#현재 resultList의 원소들이 string형태로 저장되어 있다. 이걸 딕셔너리 형태로 바꾸자.
len(resultList)
```

```python
# 웹툰이름을 key값으로 items 내용을 value로 하는 dictionary로 만든다.
for i in range(0,len(webtoonList)):
    resultDict[webtoonList[i]] = itemsList[i]
```

```python
# 2021년 연재 웹툰 대상 검색
import os
import sys
import urllib.request

client_id = "{MY_CLIENT_ID}"
client_secret = "{MY_SECRET_KEY}"

resultList = []
resultDict = {}

for webtoonName in titleList:
    encText = urllib.parse.quote("웹툰 "+webtoonName)
    #display = 100설정으로 100개 데이터 뽑아온다.
    url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "&display=100 "# json 결과 
    # url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200): # page ok
        response_body = response.read()
        #print(response_body.decode('utf-8'))
        resultList.append(response_body.decode('utf-8'))
    else:
        print("Error Code:" + rescode)
```

```python
import json
# 아이템 값만 필요하므로 'items' key를 이용해 일부분의 딕셔너리만 떼어낸다.
# 웹툰리스트의 웹툰에 모두 적용하여 2차원 리스트를 만든다.
itemsList = []
for i in range(len(resultList)):
    strToJson = json.loads(resultList[i])
    # items내용이 없는 경우가 있다. keyError발생하므로 try except를 사용한다.
    try:
        itemsList.append(strToJson['items'])
    except KeyError:
        print(titleList[i], "removed")
        del titleList[i]
        pass

itemsList[:2]
```

```python
# 일반 검색으로 네이버API가 response한 내용을 json 파일로 저장
file_path = 'data/searchItemsNaverBlog.json'
with open(file_path, 'w', encoding = "utf-8-sig") as outfile:
    outfile.write(json.dumps(resultDict, ensure_ascii=False))
```

```python
jsonList = []
# 텍스트 마이닝에 활용할 수 있는 description 부분만 떼어내서 dictionary로 저장한다.
for i in range(0,len(titleList)):
    descriptionList = []
    for j in itemsList[i]:
        #중간 중간 섞여있는 외계어들을 올바른 문자로 바꿔준다.
        description = j['description'].replace('<b>',"")
        description = description.replace('</b>',"")
        description = description.replace('&lt;',"<")
        description = description.replace('&gt;',">")
        description = description.replace('&quot;',"'")
        descriptionList.append(description)
    descriptionDict = {}
    descriptionDict['title'] = titleList[i]
    descriptionDict['blog_description'] = descriptionList
    jsonList.append(descriptionDict)
jsonList[:2]
```

```python
# title과 blog_description으로 이루어진 딕셔너리를 json 파일로 저장
file_path = 'data/blogDescription.json'
with open(file_path, 'w', encoding = "utf-8") as outfile:
    outfile.write(json.dumps(jsonList, ensure_ascii=False))
```

```python
# 이제 데이터프레임으로 바꿔서 데이터를 병합한다.
import pandas as pd
df1 = pd.read_json('data/mergeJson.json', orient = 'record', encoding = 'utf-8')
df1[:3]
```

```python
df2 = pd.read_json('data/blogDescription.json', encoding = 'utf-8')
df2[:3]
```

```python
mergeDf = pd.merge(df1, df2, how = 'inner', on = None)
mergeDf[:3]
```

```python
# 병합한 데이터프레임을 json으로 바꿔서 저장
mergeDf.to_json('data/NaverWebtoonData.json', orient = 'records', force_ascii=False)
```
